```python
from sklearn.datasets import load_iris # type: ignore
from scipy.cluster.hierarchy import dendrogram , linkage # type: ignore

n=6
X = load_iris().data[:n]
import numpy as np  # type: ignore

def getProx(X):
    mat = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            mat[i][j] = np.linalg.norm(X[i]-X[j])
    return mat

mat=getProx(X)
dendrogram(linkage(X,method='complete'))
print(mat)
```
