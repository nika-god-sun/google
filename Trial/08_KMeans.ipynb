```python
import numpy as np # type: ignore
from sklearn.datasets import load_iris # type: ignore

X = load_iris().data
y = load_iris().target

def KMeans(X,K=3):
    # print(X.shape)
    cent = X[:K]
    while True:
        exp_X=X[:,np.newaxis]
        eucDist=np.linalg.norm(exp_X-cent,axis=2)
        labels=np.argmin(eucDist,axis=1)
        newCent=np.array([X[labels==k].mean(axis=0) for k in range(K)])

        if np.all(newCent == cent):
            break
        cent = newCent
    return cent , labels

cent , labels = KMeans(X)

import matplotlib.pyplot as plt

plt.scatter(X[:,0] , X[:,1]  , c=labels)
plt.scatter(cent[:,0] , cent[:,1]  ,marker = 'x' , s = 150)
```
